```python
from pathlib import Path
Path('my_data.db').touch()

import sqlite3
conn = sqlite3.connect('my_data.db')
c = conn.cursor()

c.execute('''CREATE TABLE fishes (
                Species TEXT,Weight INTEGER,Vertical_length INTEGER ,
                Diagonal_length INTEGER,Cross_length INTEGER,Height INTEGER,
                Width INTEGER)''')

```

```python
import pandas as pd
# load the data into a Pandas DataFrame
fishes = pd.read_csv(r'Fish.csv')
# write the data to a sqlite table
fishes.to_sql('fishes', conn, if_exists='append', index = False)

c.execute('''SELECT * FROM fishes''').fetchall()

pd.read_sql('''SELECT * FROM fishes''', conn)
```

1. Company has sold too many fish everyday. Each fish has a different
weight. Company wants to know each species maximum and
minimum of each fish.

```python
#Solution:
pd.read_sql('''SELECT Species, MAX(WEIGHT),MIN(WEIGHT) FROM fishes group by Species''', conn)
```

2. The Company has many sold to much different fish and they want to
see fish which height more than 15 and also width has to be more
than 5.


```python
pd.read_sql('''SELECT Species, Height, Width FROM fishes where Height>15 and Width>5''', conn)
```

3. The company store has too many fish and each fish has its own 3
lengths. Company wants to see each species fish average lengths.

```python
pd.read_sql('''SELECT Species,AVG(Vertical_length),AVG(Diagonal_length),AVG(Cross_length) FROM fishes GROUP BY Species''', conn)
```

4. Company sees that some species sell too much. Company knows
that species of fish start at “B” and height is more than 10 and smaller
than 13. They want to all information about these fishes.

```python
pd.read_sql('''SELECT * FROM fishes WHERE Species LIKE 'B%' and Height>10 and Height<13 ''', conn)
```
